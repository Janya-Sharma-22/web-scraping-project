```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("india-news-headlines.csv")

# Check first few rows
print("Dataset preview:")
print(df.head())

# Convert publish_date to datetime
df['publish_date'] = pd.to_datetime(df['publish_date'], errors='coerce')

# --- 1. Display top 5 latest headlines ---
print("\nTop 5 Latest Headlines:")
top5 = df.sort_values(by='publish_date', ascending=False).head(5)
print(top5[['publish_date', 'headline_category', 'headline_text']])

# --- 2. Count most common categories (Trending topics) ---
category_counts = df['headline_category'].value_counts().head(10)

print("\nTop 10 Categories by Frequency:")
print(category_counts)
```

```python
#PROJECT - 02

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://realpython.github.io/fake-jobs/"
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

# Extract job info
job_cards = soup.find_all("div", class_="card-content")

jobs = []
for card in job_cards:
    title = card.find("h2", class_="title").get_text(strip=True)
    company = card.find("h3", class_="company").get_text(strip=True)
    location = card.find("p", class_="location").get_text(strip=True)
    jobs.append([title, company, location])

# Create DataFrame
df_jobs = pd.DataFrame(jobs, columns=["Job Title", "Company", "Location"])

# Save as CSV & Excel
df_jobs.to_csv("jobs.csv", index=False)
df_jobs.to_excel("jobs.xlsx", index=False)

# Basic analysis: Count jobs per city
city_counts = df_jobs["Location"].value_counts()

print("Jobs per city:")
print(city_counts)

# Show first 5 jobs
df_jobs.head()
```
